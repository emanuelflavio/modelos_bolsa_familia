```python
import seaborn as sns
import matplotlib.pyplot as plt

# Proporção de 'BOLSA_FAMILIA'
proporcao = df['BOLSA_FAMILIA'].value_counts(normalize=True)

# Gráfico de pizza
plt.pie(proporcao.values, labels=proporcao.index, autopct='%1.1f%%', colors=['green', 'red'], startangle=90, explode=(0.1, 0))
plt.title("Bolsa Família")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar o tamanho do gráfico
plt.figure(figsize=(12, 6))

# Histograma da Renda Per Capita usando Seaborn
sns.histplot(df['FAIXA_RENDA_FAMILIAR_PER_CAPITA'], bins=30, color='blue', edgecolor='black', kde=False)

# Personalizar título e eixos
plt.title("Distribuição de Renda Per Capita", fontsize=16)
plt.xlabel("Renda Per Capita (R$)", fontsize=14)
plt.ylabel("Frequência", fontsize=14)

# Grid para eixo Y
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar o gráfico
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Preparação dos dados
df_plot = df.groupby('GRAU_INSTRUCAO').size().reset_index(name='Frequência')

# Configurar o tamanho do gráfico
plt.figure(figsize=(12, 6))

# Gráfico de barras usando Seaborn
sns.barplot(data=df_plot, x='GRAU_INSTRUCAO', y='Frequência', color='red')

# Personalização
plt.title("Distribuição por Grau de Instrução", fontsize=16)
plt.xlabel("Grau de Instrução", fontsize=14)
plt.ylabel("Frequência", fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Exemplo: Criar DataFrame com os dados (substitua pelo seu arquivo real)
# df = pd.read_csv('seu_arquivo.csv')

# Agrupar os dados por sexo e bolsa família
dados_agrupados = df.groupby(['GRAU_INSTRUCAO', 'BOLSA_FAMILIA']).size().reset_index(name='CONTAGEM')

print(dados_agrupados)

# Configurar o gráfico de barras
plt.figure(figsize=(15, 10))
sns.barplot(
    data=dados_agrupados, 
    x='GRAU_INSTRUCAO', 
    y='CONTAGEM', 
    hue='BOLSA_FAMILIA', 
    palette='viridis'
)

# Adicionar título e rótulos
plt.title('Relação entre formação e Bolsa Família', fontsize=14)
plt.xlabel('GRAU_INSTRUCAO', fontsize=12)
plt.ylabel('Quantidade', fontsize=12)
plt.legend(title='Recebe Bolsa Família')
plt.tight_layout()
plt.savefig('teste1.png')

# Mostrar o gráfico
plt.show()
```
