```python
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score


df = df.drop(['PARENTESCO_RF', 'MES_ANO_REFERENCIA', 'VAL_REMUNERACAO_MES_PASSADO', 
              'DATA_NASCIMENTO', 'FAIXA_RENDA_FAMILIAR_PER_CAPITA','CRAS'], axis=1)

missing = df.isnull().sum()
print("Valores ausentes por coluna:\n", missing)




def codificar(valor):
    label_encoder = LabelEncoder()
    df[f'{valor}'] = label_encoder.fit_transform(df[f'{valor}'])
    
    for i, class_name in enumerate(label_encoder.classes_):
        print(f"{i} -> {class_name}")
    
    print('')

for i in df:
    codificar(i)

X = df.drop(['BOLSA_FAMILIA'], axis=1)
y = df['BOLSA_FAMILIA']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)



def treino(model, param_grid, model_name, X_train, y_train):
    print(f"\n=== {model_name} ===")

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    grid_search = GridSearchCV(model, param_grid, cv=cv, scoring='accuracy', n_jobs=-1, verbose=1)
    grid_search.fit(X_train, y_train)
    
    best_model = grid_search.best_estimator_
    print(f"\nMelhores Hiperparâmetros para {model_name}: {grid_search.best_params_}")
    
    cv_scores = cross_val_score(best_model, X_train, y_train, cv=cv, scoring='accuracy')
    print("\nCross-Validation Scores:")
    print(cv_scores)
    print("\nAverage CV Score:", cv_scores.mean())

    # Avaliação no conjunto de teste
    best_model.fit(X_train, y_train)
    y_pred = best_model.predict(X_test)
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    accuracy = accuracy_score(y_test, y_pred)
    print("\nAccuracy Score:", accuracy)

    return best_model


param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'criterion': ['gini', 'entropy'],
    'min_samples_split': [2, 5, 10]
}

param_grid_lr = {
    'C': [0.1, 1, 10],
    'solver': ['lbfgs', 'liblinear'],
    'penalty': ['l2']
}

param_grid_nb = {} 

param_grid_knn = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]  # Distância Manhattan (p=1) ou Euclidiana (p=2)
}




best_rf = treino(RandomForestClassifier(random_state=42), param_grid_rf, "Random Forest", X_train, y_train)

best_lr = treino(LogisticRegression(max_iter=1000, random_state=42), param_grid_lr, "Regressão Logística", X_train, y_train)


best_nb = treino(GaussianNB(), param_grid_nb, "Naive Bayes", X_train, y_train)


best_knn = treino(KNeighborsClassifier(), param_grid_knn, "KNN", X_train, y_train)


importances = best_rf.feature_importances_
importances_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': importances})
importances_df = importances_df.sort_values(by='Importance', ascending=False)

print("\nTop Features Importantes:")
print(importances_df.head(10))
```
